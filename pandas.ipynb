```python
def changer_bath(item):
    if isinstance(item,str):
        try:
            return float(item)
        except ValueError:
            return np.nan
    elif isinstance(item,float):
        return item
    else:
        return np.nan
df['NUM_BATH']=df['NUM_BATH'].apply(changer_bath)
df['NUM_BATH']
```
